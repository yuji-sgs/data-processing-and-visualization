```python
# ライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # seabornをグラフに適用させるため
```

```python
# タイタニック号のデータを扱う
df = sns.load_dataset("titanic")
```

```python
# データフレームを表示
df
```

```python
# ヒストグラムを表示
sns.displot(df["age"])
plt.show()
```

```python
# 列の各値のカウントを表示する棒グラフを生成
sns.countplot(x="pclass", data=df)
sns.catplot(x="pclass", data=df, kind="count")
plt.show()
```

```python
# 様々な変数で分けた際の平均値（今回の場合だとage）を可視化。黒い棒は95％信頼区間を示す。
sns.barplot(x="survived", y="age", hue="sex", data=df)
sns.catplot(x="survived", y="age", hue="sex", data=df, kind="bar")
plt.show()
```

```python
# 箱ひげ図を表示
sns.boxplot(x="survived", y="age", hue="sex", data=df)
sns.catplot(x="survived", y="age", hue="sex", data=df, kind="box")
plt.show()
```

```python
# ヴァイオリンプロットを表示
sns.violinplot(x="survived", y="age", hue="sex", data=df)
sns.catplot(x="survived", y="age", hue="sex", data=df, kind="violin")
plt.show()
```

```python
# ジョイントプロット（散布図とヒストグラム）を表示
sns.jointplot(x="age", y="fare", data=df)
plt.show()
```

```python
# ヒストグラムと散布図をそれぞれの変数に対応させて表示
sns.pairplot(df[["age", "fare"]])
plt.show()
```

```python
df = sns.load_dataset("iris") # 新たにirisのデータをデータセットとする
sns.pairplot(df) # 4つの量的変数に対して散布図とヒストグラムを可視化
plt.show()
```

```python
# 相関係数のヒートマップを表示
sns.heatmap(df.corr())
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing() 
df = pd.DataFrame(housing.data, columns=housing.feature_names) # 新たにCaliforniaの住宅価格データをデータセットとする

sns.heatmap(df.corr()) # 相関係数のヒートマップを表示
plt.show()
```
