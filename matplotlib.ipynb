```python
# ライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# タイタニック号のデータを扱うためseabornを使用
df = sns.load_dataset("titanic")
```

```python
# データフレームの表示
df
```

```python
# ヒストグラムを表示
plt.hist(df["fare"], bins=40, range=(0, 100))
plt.show()
```

```python
# 折れ線グラフを表示
plt.figure(figsize = (20, 5))
plt.plot(df["fare"])
plt.title("fare")
plt.xlabel("data_index")
plt.ylabel("fare")
plt.show()
```

```python
# 性別ごとの運賃のヒストグラムを表示
hist = df["fare"].hist(by = df["sex"], bins=40, range=(0, 100))
```

```python
# 複数のグラフを表示
fig, axes = plt.subplots(2, 2, figsize=(20, 10))
axes[0][0].hist(df["age"], bins=20)
axes[0][1].hist(df["fare"], bins=20)
axes[0][1].set_xlim(0, 200) # xの範囲を0〜200に指定
axes[1][0].hist(df["sex"], bins=20)
axes[1][1].hist(df["pclass"], bins=20)
plt.show()
```

```python
#df.groupby("pclass").count()["survived"]：x軸
#df.groupby("pclass").count().index：y軸

plt.bar(df.groupby("pclass").count().index, df.groupby("pclass").count()["survived"])
```

```python
# 散布図を表示
plt.scatter(df["fare"], df["age"], alpha=0.2)
plt.show()
```
