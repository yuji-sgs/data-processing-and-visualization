## **データ整形**

```python
# ageに基づいてソート（昇順）
df.sort_values('age')
```

```python
# ageに基づいてソート（降順）
df.sort_values('age',ascending=False)
```

```python
#　欠損値の合計を出力
df.isnull().sum()
```

```python
# 1つでもNaNが含まれる行だけを抽出（最初の5行のみ表示）
print(df[df.isnull().any(axis=1)].shape)
df[df.isnull().any(axis=1)].head()
```

```python
# 欠損値を削除
df.dropna()
```

```python
# 'embarkedl'列にNaNがある行を削除する
df.dropna(subset=['embarked'], axis=0, inplace=True)
df.shape # 2行削除されている
```

```python
# 欠損値確認
df['age'].head(10)
```

```python
# 欠損値を0で補完
df['age'].fillna(0).head(10)
```

```python
# 欠損値を平均値で補完
df['age'].fillna(df['age'].mean()).head(10)
```

```python
# カテゴリカル変数をダミー変数（別名ワンホットエンコーディング）に変換する操作。（ex: 男性、女性を数値の0、1で表す）
pd.get_dummies(df)
```

```python
 # 相関係数を出力
df.corr()
```
